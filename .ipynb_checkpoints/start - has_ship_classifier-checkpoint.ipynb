### create segmentation-classification trn/test dfs

```python
all_segmentations = pd.concat([train_segmentations, test_segmentations]).reset_index(drop=True)
```

```python
ntest = 20000
np.random.seed(42)
uniq_img = all_segmentations.ImageId.unique()
uniq_img = np.random.permutation(uniq_img)
trn_img = uniq_img[20000:]
test_img = uniq_img[:20000]
```

```python
len(trn_img), len(test_img)
```

```python
trn_segmentations = all_segmentations[all_segmentations.ImageId.isin(trn_img)].reset_index(drop=True)
test_segmentations = all_segmentations[all_segmentations.ImageId.isin(test_img)].reset_index(drop=True)
```

```python
trn_segmentations.shape, test_segmentations.shape
```

```python
os.makedirs(path/"segmentation", exist_ok=True)
```

```python
# move all files to train: find -type f -name "*.jpg" -exec mv -t ../train {} +
# move test images to test folder
for img in test_img:
    try:
        os.system(f"mv {path}/'train'/{img} {path}/'test'/{img}")
    except: pass
```

```python
# save segmentation csv
trn_segmentations.to_csv(path/"segmentation/trn_segmentation.csv", index=False)
test_segmentations.to_csv(path/"segmentation/test_segmentation.csv", index=False)
```

```python
# create classification csv
trn_segmentations['has_ship'] = trn_segmentations.EncodedPixels.apply(lambda x: 0 if pd.isnull(x) else 1)
test_segmentations['has_ship'] = test_segmentations.EncodedPixels.apply(lambda x: 0 if pd.isnull(x) else 1)
```

```python
trn_classification = trn_segmentations.drop("EncodedPixels", 1).drop_duplicates().reset_index(drop=True)
test_classification = test_segmentations.drop("EncodedPixels", 1).drop_duplicates().reset_index(drop=True)
```

```python
len(trn_classification), len(test_classification)
```

```python
os.makedirs(path/"classification", exist_ok=True)
```

```python
trn_classification.to_csv(path/"classification/trn_classification.csv", index=False)
test_classification.to_csv(path/"classification/test_classification.csv", index=False)
```

```python
# ratio of images without ships
trn_no_ship = (trn_classification.has_ship==0).sum() / len(trn_classification)
print(f"train images without ship: {np.round(trn_no_ship, 4)}")
```

```python
# ratio of images without ships
test_no_ship = (test_classification.has_ship==0).sum() / len(test_classification)
print(f"test images without ship: {np.round(test_no_ship, 4)}")
```

### model 1 -  has ship classifier

- set CropType.NO - not to remove ship during center cropping

```python
trn_classification = pd.read_csv(path/"classification/trn_classification.csv")
test_classification = pd.read_csv(path/"classification/test_classification.csv")
```

```python
# startified random validation set
_, val_idx = train_test_split(trn_classification, test_size=0.1, stratify=trn_classification.has_ship, random_state=42)
val_idx = (val_idx.index).tolist()
```

```python
arch = resnet18
sz = 224
```

```python
tfms = tfms_from_model(arch, sz, aug_tfms=transforms_top_down, max_zoom=1.1)
```

```python
# cv_idxs automatically creates validation set
data = ImageClassifierData.from_csv(path, folder='train', bs=512, val_idxs=val_idx,
                                    csv_fname=path/'classification/trn_classification.csv',
                                    tfms=tfms, test_name="test")
```

```python
print(f"n train: {len(data.trn_ds)}, n val: {len(data.val_ds)}")
```

```python
# number of images without ships : majority prediction
baseline = 1 - np.mean(data.val_ds.y); print(f"baseline acc: {np.round(baseline,4)}")
```

```python
torch.cuda.set_device(0)
```

```python
learn = ConvLearner.pretrained(arch, data, precompute=False)
```

```python
learn.lr_find()
```

```python
learn.sched.plot()
```

```python
learn.fit(1e-2, 10, cycle_len=1, use_clr=(20,10))
```

```python
learn.fit(1e-2, 5, cycle_len=1)
```

```python
learn.fit(1e-2, 10, cycle_len=1)
```

### unfreeze + discriminative lrs

```python
learn.unfreeze()
lr = 1e-3
lrs = [lr/100, lr/10, lr]
learn.fit(lrs, 10, cycle_len=1)
```
